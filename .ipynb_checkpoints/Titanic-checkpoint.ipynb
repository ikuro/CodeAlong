```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('titanic3.csv')
df.head()
```

## Business Understanding

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
df=df.drop(columns=['cabin','boat','body','home.dest'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='survived', hue='sex')
plt.show()
```

```python
sns.barplot(data=df, x='sex',y='age', hue='survived')
plt.show()
```

```python
sns.histplot(data=df, x='age', hue='sex',kde=True)
plt.show()
```

```python
sns.pairplot(df, hue='survived')
plt.show()
```
